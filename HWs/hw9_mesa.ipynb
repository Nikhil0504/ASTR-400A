```python
from mesa_web import read_history, read_profile
import matplotlib.pyplot as plt
```

```python
# rc params
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.size'] = 14
plt.rcParams['axes.titlesize'] = 16
plt.rcParams['axes.labelsize'] = 'x-large'
plt.rcParams['xtick.labelsize'] = 'x-large'
plt.rcParams['ytick.labelsize'] = 'x-large'
plt.rcParams['legend.fontsize'] = 14
plt.rcParams['figure.titlesize'] = 16
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 300
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['lines.markersize'] = 6
plt.rcParams['lines.markeredgewidth'] = 0.5
plt.rcParams['lines.markeredgecolor'] = 'black'
plt.rcParams['axes.grid'] = True
plt.rcParams['grid.alpha'] = 0.5
```

```python
trimmed_history_nucl = read_history('../Data/hw9_nucl_burning/trimmed_history.data', as_table=True)
```

```python
# trimmed_history_nucl.info()

# get center_h1, center_he3, center_he4, center_c12, center_o16, center_n14
center_h1 = trimmed_history_nucl['center_h1']
center_he3 = trimmed_history_nucl['center_he3']
center_he4 = trimmed_history_nucl['center_he4']
center_c12 = trimmed_history_nucl['center_c12']
center_n14 = trimmed_history_nucl['center_n14']
center_o16 = trimmed_history_nucl['center_o16']

# get time
age = trimmed_history_nucl['star_age']
```

```python
fig, ax = plt.subplots()

plt.plot(age, center_h1, label='H-1', color='blue')
# plt.plot(age, center_he3, label='He-3', color='orange')
plt.plot(age, center_he4, label='He-4', color='green')
plt.plot(age, center_c12, label='C-12', color='red')
plt.plot(age, center_n14, label='N-14', color='purple')
plt.plot(age, center_o16, label='O-16', color='brown')

# x log
plt.xscale('log')

plt.legend(loc='upper right')

plt.xlabel('Age (years)')
plt.ylabel('Mass Fraction')


# do an inset plot after 10^6 years
ax_inset = fig.add_axes([0.2, 0.7, 0.25, 0.25])

# plot the same data but zoomed in
ax_inset.plot(age, center_h1, label='H-1', color='blue')
# ax_inset.plot(age, center_he3, label='He-3', color='orange')
ax_inset.plot(age, center_he4, label='He-4', color='green')
ax_inset.plot(age, center_c12, label='C-12', color='red')
ax_inset.plot(age, center_n14, label='N-14', color='purple')
ax_inset.plot(age, center_o16, label='O-16', color='brown')

# x log
# ax_inset.set_xscale('log')

# set xlim
ax_inset.set_xlim(10**6.4, 1e7)
# ax_inset.set_xticklabels([])

# savefig
plt.savefig('ASTR-400A-HWs/figs/trimmed_history_nucl.png', dpi=300, bbox_inches='tight')
```
